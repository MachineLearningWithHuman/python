```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import plotly.offline as py
from plotly import tools
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

Firstly, we have data about the water availability in four major reservoirs that supply water to Chennai. This data spans from 2004 to 2019.

```python
df = pd.read_csv("../input/chennai_reservoir_levels.csv")
df["Date"] = pd.to_datetime(df["Date"], format='%d-%m-%Y')
df.head()
```

All the measurements are in mcft (million cubic feet).

let's plot the capacity of each resiorvior as a time series observation

```python
color=sns.color_palette()
scatter_POONDI = go.Scatter(
        x=df.Date.index[::-1],
        y=df.POONDI.values[::-1],
        showlegend=True,
        marker=dict(
            color=color,
        ),
    )
subtitles = ["Water Availability in Poondi reservoir - in mcft"]

fig = tools.make_subplots(rows=1, cols=1, vertical_spacing=0.08,
                          subplot_titles=subtitles)
fig.append_trace(scatter_POONDI, 1, 1)
fig['layout'].update(height=300, width=600, paper_bgcolor='rgb(233,233,233)')
py.iplot(fig, filename='res-plots')
```

```python
scatter_CHOLAVARAM = go.Scatter(
        x=df.Date.index[::-1],
        y=df.CHOLAVARAM.values[::-1],
        showlegend=True,
        marker=dict(
            color="brown",
        ),
    )
subtitles = ["Water Availability in CHOLAVARAM reservoir - in mcft"]

fig = tools.make_subplots(rows=1, cols=1, vertical_spacing=0.08,
                          subplot_titles=subtitles)
fig.append_trace(scatter_CHOLAVARAM, 1, 1)
fig['layout'].update(height=300, width=600, paper_bgcolor='rgb(233,233,233)')
py.iplot(fig, filename='res-plots')
```
