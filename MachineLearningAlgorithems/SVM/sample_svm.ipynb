# NoteBook For getting started with SVM

# Importing Modules

```
from sklearn.datasets.samples_generator import make_blobs
```

# Generation Of Dataset with Two classes

```
X,Y=make_blobs(n_samples=1000, centers=2,random_state=2,cluster_std=0.7)
```

here n_samples is used to denote the number of sample per cluster , centre is the cluster number random state is the randomness of number generation and cluster standard daviation is given with default value 1

# Visualization of the data

```
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=Y,s=100,cmap="spring")
plt.show()
```

visualize what will be the work of SVM classifier 

```
import numpy as np
xfit=np.linspace(-3,3)
```

```
plt.scatter(X[:,0],X[:,1],c=Y,s=100,cmap="spring")


for m,b,d in [(1,0.65,0.33),(0.5,1.6,0.55),(-0.2,2.9,0.2)]:
  yfit=m*xfit + b
  plt.plot(xfit,yfit,"-k")
  plt.fill_between(xfit,yfit-d,yfit+d,edgecolor="none" ,alpha=0.4)
plt.xlim(-3,3)
plt.show()
```

definetly not a good classifier but to give you a concept
